### 1.Find those groups which have more “True” values than “False” values in the below dataframe:

```python
df = pd.DataFrame({'A': ['group1', 'group1', 'group2', 'group1','group2','group1', 'group2', 'group2','group2','group1'],'B': ['true', 'true', 'true', 'false', 'false', 'false', 'false', 'true','false','true']})
```

```python
df
```

```python
true=df[df['B']=='true'].groupby('A').count()     #grouping for true values                     
false=df[df['B']=='false'].groupby('A').count()   #grouping for False values

for i in true.index.values:                      
    if true.loc[i,'B']>false.loc[i,'B']:          #comparing
        print(i)
```

### 3.a.Generate a random series of length 10 and find the positions of numbers that are multiples of 3 from a series?

```python
randomSeries=pd.Series(np.random.randint(1,101,size=10))                   #Generating a Random Series of Length 10

print(f"Generated Random Series:\n{randomSeries}")
print("\nIndex of Numbers that are Multiple of 3:")
for i in randomSeries.index.values:                                        #Iterating through the index of the Series
    if randomSeries[i]%3==0:
        print(i)
```

### 5.Find the desired output as shown in the image below.
**Note**: Create the data frame (Train Data) as shown in the image below (Output can be in fraction form or decimal form )

```python
train_data=pd.DataFrame({'State':['A','B','C','A','A','B','A','A','C','C'],
                         'Class':[0,1,1,0,1,1,0,1,1,0]})                     #Creating train_data Data Frame
train_data
```

```python
#Creating Response Table from Train Data
responseTable=train_data.groupby(['State','Class']).size().unstack()
responseTable[0].fillna(0,inplace=True)
responseTable['Sum']=responseTable[0]+responseTable[1]
responseTable
```

```python
#Producing Desired Output
encoded_train_data=pd.DataFrame()
encoded_train_data['class_0']=train_data['State'].map(responseTable[0])/train_data['State'].map(responseTable['Sum'])
encoded_train_data['class_1']=train_data['State'].map(responseTable[1])/train_data['State'].map(responseTable['Sum'])
encoded_train_data
```
